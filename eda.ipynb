# California Fire Data Analysis

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('California Wildfire Damage.csv')
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
print(df.info())
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df['Area_Burned (Acres)'], bins=30, kde=True)
plt.title('Distribution of Area Burned (Acres)')
plt.show()
```

```python
df.columns
```

```python
plt.figure(figsize=(10,5))
sns.countplot(y=df['Cause'], order=df['Cause'].value_counts().index)
plt.title('Most Common Causes of Wildfires')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=df[['Injuries', 'Fatalities']])
plt.title('Distribution of Injuries and Fatalities')
plt.show()
```

```python
plt.figure(figsize=(12,6))
df['Month'] = df['Date'].dt.month
df_monthly = df.groupby('Month').size()
sns.lineplot(x=df_monthly.index, y=df_monthly.values, marker='o')
plt.title('Monthly Trend of Wildfires')
plt.xlabel('Month')
plt.ylabel('Number of Incidents')
plt.show()
```

```python
top_locations = df['Location'].value_counts().head(10)
plt.figure(figsize=(10,5))
sns.barplot(x=top_locations.index, y=top_locations.values)
plt.xticks(rotation=45)
plt.title('Top 10 Most Affected Locations')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(data=df, x='Homes_Destroyed', y='Businesses_Destroyed', hue='Cause', size='Fatalities', sizes=(20,200))
plt.title('Impact of Wildfires on Homes & Businesses')
plt.show()
```

```python
df.drop(labels=['Incident_ID','Location','Cause'],axis=1,inplace=True)    
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Feature Correlation Heatmap')
plt.show()
```
